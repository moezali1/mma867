```python
data = pd.read_csv("AirPassengers.csv")
data['Month'] = pd.to_datetime(data['Month'])  # Convert to datetime
data.set_index('Month', inplace=True)  # Set datetime as index
```

```python
import matplotlib.pyplot as plt

# Plotting the seasonal plot
plt.figure(figsize=(10, 5))
for year, group in data.groupby(data.index.year):
    plt.plot(group.index.month, group['#Passengers'], label=year)

plt.title('Seasonal Plot of Airline Passengers')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend(title='Year')
plt.grid(True)
plt.show()
```
